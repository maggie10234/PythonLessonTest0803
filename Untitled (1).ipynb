```python
#秒數變時間
from datetime import datetime
datetime.utcfromtimestamp(1140825600)
str(datetime.utcfromtimestamp(1140825600))
```

```python
def timestrans(filmdate):
    return str(datetime.utcfromtimestamp(filmdate))
# apply 歷遍
df["film_date"].apply(timestrans)
```

```python
# 做完的加到新欄位
df["real_film_date"] = df["film_date"].apply(timestrans)
df
```

```python
# True的才會顯示出來
# 參數數量要和取出的數量一樣
# "[True, False, True]" 這是一個list
df.head(3)[[True, False, True]]
```

```python
# 篩選結果
def fflow(n):
    if n > 3000 and n < 5000:
        return True
    else:
        return False
f = df["comments"].apply(fflow)
df[f]
```

```python
a = eval("[1, 2, 3]")
a
```

```python
# eval 可以把字串當成程式碼直接執行
def fflow(mm):
    tagList = eval(mm)
    if "children" in tagList:
        return True
    else:
        return False
df[ df["tags"].apply(fflow) ]
# 外面加上"df[    ]"才會印出來
```
